```python
uno = pd.read_csv('./outputs/Training_comparativos/corrida_may_jun.csv')
uno
```

```python
dos = pd.read_csv('./outputs/Training_comparativos/corrida_jul_ago.csv')
tres = pd.read_csv('./outputs/Training_comparativos/corrida_sep_oct.csv')
cuatro = pd.read_csv('./outputs/Training_comparativos/corrida_nov_dic.csv')
```

```python
retraining = pd.concat([uno, dos])
retraining = pd.concat([retraining, tres])
retraining = pd.concat([retraining, cuatro])
retraining.rename(columns={'MAPE': 'MAPE_rt', 'R2':'R2_rt'}, inplace=True)
retraining
```

```python
original = pd.merge(original, retraining[['FECHA-HORA', 'MAPE_rt', 'R2_rt']], on='FECHA-HORA', how='left')
original
```
